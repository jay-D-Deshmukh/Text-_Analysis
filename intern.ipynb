```python
from nltk.corpus import stopwords
import requests
from bs4 import BeautifulSoup
from textblob import TextBlob
import nltk
from nltk import sent_tokenize, word_tokenize
import syllables



# Download the NLTK resources (if not already downloaded)
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

# Function to get text content from a URL
def get_text_from_url(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    return soup.get_text()

def avg_sentence_length(text):
    sentences = sent_tokenize(text)
    total_words = sum(len(word_tokenize(sentence)) for sentence in sentences if sentence.strip())
    total_sentences = len(sentences)
    return total_words / total_sentences if total_sentences > 0 else 0



# Function to calculate average word per sentence
def avg_words_per_sentence(text):
    sentences = sent_tokenize(text)
    total_words = sum(len(word_tokenize(sentence)) for sentence in sentences)
    total_sentences = len(sentences)
    return total_words / total_sentences

# Function to calculate percentage of complex words
def percentage_complex_words(text):
    words = word_tokenize(text)
    tagged = nltk.pos_tag(words)
    complex_words = [word for word, tag in tagged if tag.startswith('NN') or tag.startswith('VB')]
    return (len(complex_words) / len(words)) * 100

def complex_words(text):
    words = word_tokenize(text)
    tagged = nltk.pos_tag(words)
    complex_words = [word for word, tag in tagged if tag.startswith('NN') or tag.startswith('VB')]
    return len(complex_words)
```

```python
def count_syllables(word):
    # Handling exceptions for words ending with "es" and "ed"
    exceptions = ["es", "ed"]
    if word[-2:] in exceptions:
        return 0

    # Counting vowels to estimate syllable count
    vowels = "aeiouAEIOU"
    count = 0
    prev_char = ''

    for char in word:
        if char in vowels and prev_char not in vowels:
            count += 1
        prev_char = char

    # Handling special case for words ending with "e"
    if word[-1] == 'e' and word[-2] not in vowels:
        count -= 1

    return max(1, count)  # Ensure at least 1 syllable

def count_syllables_in_text(text):
    words = text.split()
    return [count_syllables(word) for word in words]
```

```python
import re

def count_personal_pronouns(text):
    # Define a regex pattern to match personal pronouns
    pronoun_pattern = r'\b(I|we|my|ours|us)\b'

    # Find all matches of the pattern in the text
    matches = re.findall(pronoun_pattern, text, flags=re.IGNORECASE)

    return len(matches)
```

```python
# List of links to scrape
links = url_link
```

```python
results = []

for link in links:
    text = get_text_from_url(link)
    analysis = TextBlob(text)

    result = {
        'Link': link,
        'Positive Score': analysis.sentiment.polarity,
        'Negative Score': -analysis.sentiment.polarity,
        'Polarity Score': analysis.sentiment.polarity,
        'Subjectivity Score': analysis.sentiment.subjectivity,
        'Avg Sentence Length': avg_sentence_length(text),
        'Percentage of Complex Words': percentage_complex_words(text),
        'FOG Index': 0.4 * (avg_words_per_sentence(text) + percentage_complex_words(text)),
        'Avg Number of Words Per Sentence': avg_words_per_sentence(text),
        'Complex Word Count': complex_words(text),
        'Word Count': len(word_tokenize(text)),
        'Syllable Per Word':count_syllables_in_text(text),
        'Personal Pronouns':count_personal_pronouns(text),
        'Avg Word Length': sum(len(word) for word in word_tokenize(text)) / len(word_tokenize(text))
    }

    results.append(result)

# Importing pandas for data manipulation
import pandas as pd

# Creating a DataFrame from the results
df = pd.DataFrame(results)

# Saving the DataFrame to a CSV file
df.to_csv('text_data_metrics.csv', index=False)
```
